# Character RNN

```python
import tensorflow as tf
from tensorflow import keras
```

```python
shakespeare_url = "https://homl.info/shakespeare" # shortcut URL 
filepath = keras.utils.get_file("shakespeare.txt", shakespeare_url) 
with open(filepath) as f:
    shakespeare_text = f.read()
```

```python
len(shakespeare_text)
```

```python
# convert characters to numbers - Tokenizer class
# char_level = True to get character level encoding
tokenizer = keras.preprocessing.text.Tokenizer(char_level=True)
tokenizer.fit_on_texts([shakespeare_text])
```

```python
# Converting characters to numbers.
tokenizer.texts_to_sequences(["First"])
```

```python
# Converting numbers to characters.
tokenizer.sequences_to_texts([[20, 6, 9, 8, 3]])
```

```python
max_id = len(tokenizer.word_index)
```

```python
max_id # no.of.distinct characters.
```

```python
dataset_size = 1115394 # total number of characters
```

```python
train_size = dataset_size * 90 // 100
```

```python
import numpy as np
```

```python
[encoded] = np.array(tokenizer.texts_to_sequences([shakespeare_text])) - 1
```

```python
len(encoded)
```

```python
dataset_size * 90 // 100
```

```python
# spliting training data - 90 % 
# Using the tf data set API to get one character at a time from the tensor.

train_size = dataset_size * 90 // 100
dataset = tf.data.Dataset.from_tensor_slices(encoded[:train_size])
```

## Chopping the Sequential Dataset into Multiple Windows

```python
# Window method
# by default it is nonoverlapping, using shift =1 gives more training data. 
n_steps = 100
window_length = n_steps + 1 # target = input shifted 1 character ahead 
dataset = dataset.window(window_length, shift=1, drop_remainder=True)
```

```python
dataset
```

```python
for window in dataset:
    print(list(window.as_numpy_iterator()))
```

```python
dataset # now it is a nested dataset - DatasetSpecc object
```

```python
# Using flat_map to convert the datasets to tensorflow.
# we can pass any function in the flat_map to perform transformation as well.

dataset = dataset.flat_map(lambda window: window.batch(window_length))
```

```python
dataset.shuffle(10000).batch(batch_size)
```

```python
# Shuffling the data - Since GD will work better. 

batch_size = 32
dataset = dataset.shuffle(10000).batch(batch_size)
dataset = dataset.map(lambda windows: (windows[:, :-1], windows[:, 1:]))
```

```python
dataset
```

```python
dataset = dataset.map(lambda X_batch, Y_batch: (tf.one_hot(X_batch, depth=max_id), Y_batch))
```

```python
dataset = dataset.prefetch(1)
```

```python
# Create a Char RNN model
model = keras.models.Sequential([
    keras.layers.GRU(128, return_sequences=True, input_shape=[None, max_id],
                     dropout=0.2, recurrent_dropout=0.2),
    keras.layers.GRU(128, return_sequences=True,
                     dropout=0.2, recurrent_dropout=0.2),
    keras.layers.TimeDistributed(keras.layers.Dense(max_id,
                                                activation="softmax"))])
```

```python
model.compile(loss="sparse_categorical_crossentropy", optimizer="adam")

history = model.fit(dataset, epochs=5)
```

```python
# Predicting using the Char-RNN Model

def preprocess(texts):
    X = np.array(tokenizer.texts_to_sequences(texts)) - 1 
    return tf.one_hot(X, max_id)
```

```python
X_new = preprocess(["How are yo"])
Y_pred = model.predict_classes(X_new)
tokenizer.sequences_to_texts(Y_pred + 1)[0][-1] # 1st sentence, last char
```
